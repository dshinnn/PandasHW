### Homework Excerise #1 - White Sox Data<br>
<p>Return a dataframe showing the White Sox qualified hitters* in order from best to worst XBHA**.</p>
<p>*A hitter must have at least 100 At Bats (AB's) to qualify</p>
<p>** XBHA is calculated as the number of extra base hits (2B, 3B, HR) divided by the number of At Bats (AB's)</p>

```python
import pandas as pd
```

```python
df = pd.read_csv('whitesox_2021_hitting.csv')
df
```

```python
# Remove disqualified atheletes 
qualified = df.where(df['AB'] >= 100).dropna()
qualified
```

```python
# Performs calculation for XBHA
def calc_xbha(row):
    return (row['2B'] + row['3B'] + row['HR']) / row['AB']

# Applies the function to each row and assign result to new XBHA column
qualified['XBHA'] = qualified.apply(calc_xbha, axis=1)
# Filters dataframe with essential columns only
qualified[['Pos', 'Name', '2B', '3B', 'HR', 'AB', 'XBHA']].sort_values('XBHA', ascending=False).reset_index(drop=True)
```
